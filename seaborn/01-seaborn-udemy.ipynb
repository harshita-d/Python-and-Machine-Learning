# Seaborn

### * Distribution Plots *

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

#### ** DIST  Plot** 

allows us to show distribution of univariate set of observations 

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'],kde=False)
#kde=False removes the line
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
#try increasing value of bins it will increase the total number of histogram 
#if bin=30 than total number of histograms =30
```

#### ** JOINT  Plot** 

to basically match up to this plots for bi- variate data meaning you can essentially combine two different distribution plots.

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#darker color means it has many points at that place and it fades as the total number of points at that place goes on decreasing
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#it draws a regression line
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

#### ** PAIR  Plot** 

to plot pairwise relationships across an entire data frame at least for the numerical columns.

And it also supports a color hue argument for categorical columns

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='husl')
```

#### ** RUG  Plot** 

so the difference between a histogram here below and this Rugge plot is that the histogram essentially has bins and it counts how many dashes were in that bin and then shows it as a number up on y axis
ex: 45 dashes there between 10 to 11. They're all kind of stacked on top of each other.the histogram is going to be as high as 45.

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

#### KDE:- kernel density estimation

```python
sns.kdeplot(tips['total_bill'])
```

### Categorical Plots

#### ** Bar Plots **

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#it gives the average or mean of total bill for each sex
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

#### ** Count Plot **

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

#### ** Strip Plot **

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
#jitter adds random noise which is by default True
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

#### ** Swarm Plot **

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
#it is a mixture of swarm as well as strip plot
#means no two points are overlapped
```

#### ** factor plot **

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
# in kind we can write the name of different plots used
```

### Matrix Plots

Your data should already be in a matrix form and what it mean by matrix form is so that the index name and the column name match up so that the cell value actually indicates something that is relevant to both of those names.

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc=tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
flights
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp,linecolor='white',linewidths=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
#months and years are arranged randomly according to their similarity
#like 1959 and 1960 are similar and june and september are similar
```

### Grid

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

#### ** Pair Plots **

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g=sns.PairGrid(iris)
g.map(plt.bar)
#it generates an empty grid
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

#### ** Facet Grid **

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
#for scatter plot we need two argument
```

#### ** Lm Plot **

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['v','*'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['v','*'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=8)
```
