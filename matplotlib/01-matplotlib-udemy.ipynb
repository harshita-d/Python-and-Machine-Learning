### Matplotlib-1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#this line and that's going to allow you to actually see the plots you create inside the Jupiter notebook
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

#### *****functional method*****

```python
plt.plot(x,y,)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
#multiplot on same surface
plt.subplot(1,2,1)#rows,columns
plt.plot(x,y,'r')

plt.subplot(1,2,2)#rows,columns
plt.plot(y,x,'b')
```

#### *****Object Oriented method*****

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#axes can be added through a set of lists
#left bottom width and height arguments that range from 0 to 1.
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
axes[0].plot(x,x**2,x,np.exp(x))
axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig,ax=plt.subplots(figsize=(10,4))
ax.plot(x,x**2,x,x**3,lw=2)

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(['alpha', 'beta', 'gamma', 'delta', 'epsilon'],fontsize=18)

ax.set_yticks([0,50,100,150])
```

```python
fig,ax=plt.subplots(figsize=(10,4))
ax.plot(x,x**2,x,x**3,lw=2)

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'],fontsize=18)

yticks=[0,50,100,150]
ax.set_yticks([0,50,100,150])
ax.set_yticklabels(['$%.1f$' % y for y in yticks],fontsize=18)
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```

```python
import matplotlib
matplotlib.rcParams['xtick.major.pad'] = 20
matplotlib.rcParams['ytick.major.pad'] = 10

fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("label and axis spacing")

# padding between axis label and axis numbers
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 10

ax.set_xlabel("x")
ax.set_ylabel("y");
```
